```python
import glob
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Reshape, LSTM, Dense
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# 1. Load Dataset
# ===============================
bengin_cases = glob.glob(r"D:\Downloads\LUNG_NEW_EXTRACTEDFILE\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Bengin cases\*.*")
malign_cases = glob.glob(r"D:\Downloads\LUNG_NEW_EXTRACTEDFILE\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Malignant cases\*.*")
normal_cases = glob.glob(r"D:\Downloads\LUNG_NEW_EXTRACTEDFILE\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Normal cases\*.*")

print("Benign:", len(bengin_cases))
print("Malignant:", len(malign_cases))
print("Normal:", len(normal_cases))

images = []
labels = []

for img_path in bengin_cases:
    image = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    image = img_to_array(image)
    images.append(image)
    labels.append(0)  # Benign

for img_path in malign_cases:
    image = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    image = img_to_array(image)
    images.append(image)
    labels.append(1)  # Malignant

for img_path in normal_cases:
    image = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    image = img_to_array(image)
    images.append(image)
    labels.append(2)  # Normal

images = np.array(images, dtype='float32') / 255.0
labels = np.array(labels)

# ===============================
# 2. Train-Test Split
# ===============================
train_x, test_x, train_y, test_y = train_test_split(images, labels, test_size=0.2, random_state=42)

# ===============================
# 3. CNN + LSTM Hybrid Model
# ===============================
inputs = Input(shape=(224, 224, 3))
x = Conv2D(32, kernel_size=3, activation='relu', kernel_initializer='he_normal')(inputs)
x = MaxPooling2D(2, 2)(x)  # Output: (112, 112, 32)
x = Conv2D(32, kernel_size=3, activation='relu', kernel_initializer='he_normal')(x)
x = MaxPooling2D(2, 2)(x)  # Output: (55, 55, 32)
x = Reshape((-1, 32))(x)   # Shape: (batch, time_steps, features)
x = LSTM(64)(x)
x = Dense(32, activation='elu', kernel_initializer='he_uniform')(x)
outputs = Dense(3, activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)

# ===============================
# 4. Compile and Train
# ===============================
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(train_x, train_y, epochs=40, validation_split=0.2)

# Save the trained model
model.save('model.h5')
print("Model saved as model.h5")

# ===============================
# 5. Evaluation
# ===============================
print("\nEvaluating on test data...")
test_loss, test_accuracy = model.evaluate(test_x, test_y)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Predictions
pred_y = np.argmax(model.predict(test_x), axis=1)

# Classification Report
print("\nClassification Report:")
print(classification_report(test_y, pred_y, target_names=['Benign', 'Malignant', 'Normal']))

# Confusion Matrix
cm = confusion_matrix(test_y, pred_y)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant', 'Normal'], yticklabels=['Benign', 'Malignant', 'Normal'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```

```python
import glob
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Flatten, Dense,
    LayerNormalization, MultiHeadAttention, Dropout,
    Add, Reshape, GlobalAveragePooling1D
)

# ===============================
# 1. Load Dataset
# ===============================
bengin_cases = glob.glob(r"D:\Downloads\LUNG_NEW_EXTRACTEDFILE\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Bengin cases\*.*")
malign_cases = glob.glob(r"D:\Downloads\LUNG_NEW_EXTRACTEDFILE\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Malignant cases\*.*")
normal_cases = glob.glob(r"D:\Downloads\LUNG_NEW_EXTRACTEDFILE\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Normal cases\*.*")

images = []
labels = []

# Load and preprocess images
for img_path in bengin_cases:
    image = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    image = img_to_array(image)
    images.append(image)
    labels.append(0)  # Bengin

for img_path in malign_cases:
    image = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    image = img_to_array(image)
    images.append(image)
    labels.append(1)  # Malignant

for img_path in normal_cases:
    image = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    image = img_to_array(image)
    images.append(image)
    labels.append(2)  # Normal

# Convert to NumPy arrays
images = np.array(images, dtype='float32') / 255.0
labels = np.array(labels)

# ===============================
# 2. Train-Test Split
# ===============================
train_x, test_x, train_y, test_y = train_test_split(images, labels, test_size=0.2, random_state=42)

# ===============================
# 3. Transformer Encoder Block
# ===============================
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    x = LayerNormalization(epsilon=1e-6)(inputs)
    x = MultiHeadAttention(num_heads=num_heads, key_dim=head_size, dropout=dropout)(x, x)
    x = Dropout(dropout)(x)
    res = Add()([x, inputs])

    x = LayerNormalization(epsilon=1e-6)(res)
    x = Dense(ff_dim, activation='relu')(x)
    x = Dropout(dropout)(x)
    x = Dense(inputs.shape[-1])(x)
    return Add()([x, res])

# ===============================
# 4. CNN + Transformer Model
# ===============================
inputs = Input(shape=(224, 224, 3))
x = Conv2D(32, kernel_size=3, activation='relu', kernel_initializer='he_normal')(inputs)
x = MaxPooling2D(2, 2)(x)
x = Conv2D(32, kernel_size=3, activation='relu', kernel_initializer='he_normal')(x)
x = MaxPooling2D(2, 2)(x)
x = Flatten()(x)

# Reshape to sequence (144, 648) for Transformer (since 224x224x3 -> downsampled and flattened to 93312)
x = Reshape((144, 648))(x)

# Add Transformer
x = transformer_encoder(x, head_size=32, num_heads=2, ff_dim=64, dropout=0.1)

# Global pooling and output
x = GlobalAveragePooling1D(name='transformer_output')(x)
outputs = Dense(3, activation='softmax')(x)

model = Model(inputs=inputs, outputs=outputs)
model.summary()

# ===============================
# 5. Compile and Train
# ===============================
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_x, train_y, epochs=3, validation_split=0.2)

# ===============================
# 6. Extract Transformer Output
# ===============================
intermediate_model = Model(inputs=model.input, outputs=model.get_layer('transformer_output').output)
sample_input = np.expand_dims(train_x[0], axis=0)  # add batch dimension
transformer_output = intermediate_model.predict(sample_input)

print("Transformer output shape:", transformer_output.shape)
print("Transformer output values:", transformer_output)
```

```python
import os
import glob
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ===============================
# 1. Load Dataset
# ===============================
bengin_cases = glob.glob(r"D:\Downloads\LUNG_NEW_extract\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Bengin cases\*.*")
malign_cases = glob.glob(r"D:\Downloads\LUNG_NEW_extract\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Malignant cases\*.*")
normal_cases = glob.glob(r"D:\Downloads\LUNG_NEW_extract\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Normal cases\*.*")

images, labels = [], []

for img_path in bengin_cases:
    img = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    images.append(img_to_array(img))
    labels.append(0)

for img_path in malign_cases:
    img = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    images.append(img_to_array(img))
    labels.append(1)

for img_path in normal_cases:
    img = load_img(img_path, color_mode='rgb', target_size=(224, 224))
    images.append(img_to_array(img))
    labels.append(2)

images = np.array(images, dtype='float32') / 255.0
labels = np.array(labels)
print(f"Total images loaded: {len(images)}")

# ===============================
# 2. Train-Test Split
# ===============================
train_x, test_x, train_y, test_y = train_test_split(
    images, labels, test_size=0.2, stratify=labels, random_state=42
)

# ===============================
# 3. Data Augmentation
# ===============================
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
])

# ===============================
# 4. Build EfficientNet Model (with fine-tuning)
# ===============================
base_model = EfficientNetB0(include_top=False, weights='imagenet', input_shape=(224, 224, 3), pooling='avg')
base_model.trainable = True  # Fine-tune entire model

inputs = tf.keras.Input(shape=(224, 224, 3))
x = data_augmentation(inputs)
x = base_model(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(3, activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)

# ===============================
# 5. Compile & Train Model
# ===============================
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-6)

history = model.fit(
    train_x, train_y,
    validation_split=0.2,
    epochs=40,
    batch_size=32,
    callbacks=[reduce_lr]
)

# ===============================
# 6. Evaluate on Test Data
# ===============================
test_loss, test_acc = model.evaluate(test_x, test_y)
print("Test accuracy:", test_acc)
```

```python
import tensorflow as tf
import numpy as np
import glob
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ===============================
# Custom LoRA Layer
# ===============================
class LoRALayer(tf.keras.layers.Layer):
    def __init__(self, units, r=12, alpha=1.0, **kwargs):
        super(LoRALayer, self).__init__(**kwargs)
        self.units = units
        self.r = r
        self.alpha = alpha

    def build(self, input_shape):
        input_dim = input_shape[-1]

        # Frozen weight (D)
        self.D = self.add_weight(
            shape=(input_dim, self.units),
            initializer="glorot_uniform",
            trainable=False,
            name="frozen_weight"
        )

        # LoRA adapters: K (A), R (B)
        self.A = self.add_weight(
            shape=(self.r, input_dim),
            initializer="random_normal",
            trainable=True,
            name="lora_A"
        )
        self.B = self.add_weight(
            shape=(self.units, self.r),
            initializer="random_normal",
            trainable=True,
            name="lora_B"
        )

    def call(self, inputs):
        original = tf.matmul(inputs, self.D)
        lora_update = tf.matmul(inputs, tf.transpose(self.A))  # (batch, r)
        lora_update = tf.matmul(lora_update, tf.transpose(self.B))  # (batch, units)
        return original + self.alpha * lora_update

# ===============================
# 1. Load Dataset
# ===============================
bengin_cases = glob.glob(r"D:\Downloads\LUNG_NEW_extract\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Bengin cases\*.*")
malign_cases = glob.glob(r"D:\Downloads\LUNG_NEW_extract\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Malignant cases\*.*")
normal_cases = glob.glob(r"D:\Downloads\LUNG_NEW_extract\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset\Normal cases\*.*")

images, labels = [], []

def load_and_preprocess(image_path):
    image = load_img(image_path, color_mode='rgb', target_size=(224, 224))
    image = img_to_array(image)
    return image

for img_path in bengin_cases:
    images.append(load_and_preprocess(img_path))
    labels.append(0)

for img_path in malign_cases:
    images.append(load_and_preprocess(img_path))
    labels.append(1)

for img_path in normal_cases:
    images.append(load_and_preprocess(img_path))
    labels.append(2)

images = np.array(images, dtype='float32') / 255.0
labels = np.array(labels)

# ===============================
# 2. Train-Test Split
# ===============================
train_x, test_x, train_y, test_y = train_test_split(images, labels, test_size=0.2, random_state=42)

# ===============================
# 3. Data Augmentation
# ===============================
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_gen = datagen.flow(train_x, train_y, batch_size=32, subset='training')
val_gen = datagen.flow(train_x, train_y, batch_size=32, subset='validation')

# ===============================
# 4. Load Pretrained EfficientNet
# ===============================
base_model = tf.keras.applications.EfficientNetB0(
    include_top=False,
    weights='imagenet',
    input_shape=(224, 224, 3),
    pooling='avg'
)

# Freeze all layers
for layer in base_model.layers:
    layer.trainable = False

# Unfreeze last 30 layers (partial fine-tuning)
for layer in base_model.layers[-30:]:
    layer.trainable = True

# ===============================
# 5. Model with LoRA Head
# ===============================
inputs = tf.keras.Input(shape=(224, 224, 3))
x = base_model(inputs, training=True)
x = tf.keras.layers.BatchNormalization()(x)
x = LoRALayer(units=128, r=8, alpha=1.0)(x)  # LoRA applied here
x = tf.keras.layers.ReLU()(x)
outputs = tf.keras.layers.Dense(3)(x)  # 3 classes (logits)

model = tf.keras.Model(inputs, outputs)

# ===============================
# 6. Compile Model
# ===============================
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

# ===============================
# 7. Train Model
# ===============================
model.fit(train_gen, validation_data=val_gen, epochs=10)

# ===============================
# 8. Evaluate Model
# ===============================
test_loss, test_acc = model.evaluate(test_x, test_y)
print("Test accuracy:", test_acc)
```

```python
import tensorflow as tf
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# ===========================================
# 1. Enhanced LoRA Implementation
# ===========================================
class LoRALayer(layers.Layer):
    def __init__(self, original_layer, r=16, alpha=2.0, **kwargs):
        super(LoRALayer, self).__init__(**kwargs)
        self.r = r
        self.alpha = alpha
        
        # Get original layer weights
        original_weights = original_layer.get_weights()
        self.original_kernel = tf.Variable(
            original_weights[0], 
            trainable=False,
            name="original_kernel"
        )
        
        # Initialize LoRA matrices
        self.A = self.add_weight(
            shape=(self.original_kernel.shape[0], r),
            initializer="random_normal",
            trainable=True,
            name="lora_A"
        )
        self.B = self.add_weight(
            shape=(r, self.original_kernel.shape[1]),
            initializer="zeros",
            trainable=True,
            name="lora_B"
        )
        
    def call(self, inputs):
        original_output = tf.matmul(inputs, self.original_kernel)
        lora_output = tf.matmul(tf.matmul(inputs, self.A), self.B)
        return original_output + (self.alpha * lora_output)

# ===========================================
# 2. Data Loading with Verification
# ===========================================
def load_dataset():
    base_path = r"D:\Downloads\LUNG_NEW_extract\The IQ-OTHNCCD lung cancer dataset\The IQ-OTHNCCD lung cancer dataset"
    paths = {
        'benign': os.path.join(base_path, "Bengin cases", "*.*"),
        'malignant': os.path.join(base_path, "Malignant cases", "*.*"),
        'normal': os.path.join(base_path, "Normal cases", "*.*")
    }

    images, labels = [], []
    class_names = ['benign', 'malignant', 'normal']
    
    for i, (cls, path) in enumerate(paths.items()):
        print(f"Loading {cls} cases...")
        for img_path in glob.glob(path):
            try:
                img = load_img(img_path, target_size=(224, 224))
                img = img_to_array(img)
                images.append(img)
                labels.append(i)
            except Exception as e:
                print(f"Error loading {img_path}: {str(e)}")
    
    images = np.array(images, dtype='float32') / 255.0
    labels = np.array(labels)
    
    # Verify dataset
    print("\nDataset Summary:")
    for i, name in enumerate(class_names):
        print(f"{name}: {np.sum(labels == i)} samples")
    
    if len(images) == 0:
        raise ValueError("No images were loaded. Check your paths and file formats.")
    
    return images, labels, class_names

# ===========================================
# 3. Data Visualization
# ===========================================
def visualize_data(images, labels, class_names):
    plt.figure(figsize=(12, 8))
    for i in range(12):
        plt.subplot(3, 4, i+1)
        plt.imshow(images[i])
        plt.title(class_names[labels[i]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# ===========================================
# 4. Model Construction
# ===========================================
def build_model():
    # Load base model
    base_model = tf.keras.applications.EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_shape=(224, 224, 3),
        pooling='avg'
    )
    
    # Freeze base model
    base_model.trainable = False
    
    # Create LoRA dense layer
    original_dense = layers.Dense(1280)
    original_dense.build((None, 1280))  # Match EfficientNet output shape
    
    # Build model
    inputs = layers.Input(shape=(224, 224, 3))
    x = base_model(inputs)
    x = LoRALayer(original_dense, r=16, alpha=2.0)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Dropout(0.5)(x)
    
    # Classifier head
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(256, activation='relu')(x)
    outputs = layers.Dense(3, activation='softmax')(x)
    
    model = models.Model(inputs, outputs)
    
    # Gradual unfreezing
    for layer in base_model.layers[-30:]:
        layer.trainable = True
    
    return model

# ===========================================
# 5. Training Pipeline
# ===========================================
def train_model():
    # Load data
    images, labels, class_names = load_dataset()
    visualize_data(images, labels, class_names)
    
    # Split data
    train_x, test_x, train_y, test_y = train_test_split(
        images, labels, 
        test_size=0.2, 
        random_state=42,
        stratify=labels
    )
    
    # Data augmentation
    train_datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.3,
        horizontal_flip=True,
        vertical_flip=True,
        brightness_range=[0.7, 1.3],
        validation_split=0.2
    )
    
    train_gen = train_datagen.flow(
        train_x, train_y,
        batch_size=32,
        subset='training',
        shuffle=True
    )
    
    val_gen = train_datagen.flow(
        train_x, train_y,
        batch_size=32,
        subset='validation',
        shuffle=False
    )
    
    # Build and compile model
    model = build_model()
    
    # Class weights
    class_weights = class_weight.compute_class_weight(
        'balanced',
        classes=np.unique(train_y),
        y=train_y
    )
    class_weights = dict(enumerate(class_weights))
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Define callbacks (fixed the naming conflict)
    training_callbacks = [
        EarlyStopping(patience=10, restore_best_weights=True, monitor='val_accuracy'),
        ReduceLROnPlateau(factor=0.5, patience=3, monitor='val_loss'),
        ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_accuracy')
    ]
    
    # Train
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=50,
        callbacks=training_callbacks,
        class_weight=class_weights
    )
    
    # Evaluate
    test_loss, test_acc = model.evaluate(test_x, test_y)
    print(f"\nFinal Test Accuracy: {test_acc:.4f}")
    
    # Plot training history
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return model

# Run training
model = train_model()
```

```python
import os
import shutil
import random
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === Paths ===
base_dir = r"D:\Downloads\CT_SCAN_LUNG_NODULES_EXTRACTEDFILE\data"
train_dir = os.path.join(base_dir, "train")
val_dir = os.path.join(base_dir, "val")
test_dir = os.path.join(base_dir, "test")

# === Auto create val folder if missing ===
def create_validation_split(train_dir, val_dir, split_ratio=0.2):
    if os.path.exists(val_dir):
        print("[INFO] Validation directory already exists.")
        return
    os.makedirs(val_dir)
    for cls in os.listdir(train_dir):
        class_train = os.path.join(train_dir, cls)
        class_val = os.path.join(val_dir, cls)
        os.makedirs(class_val, exist_ok=True)
        images = os.listdir(class_train)
        val_samples = random.sample(images, int(split_ratio * len(images)))
        for img in val_samples:
            shutil.move(os.path.join(class_train, img), os.path.join(class_val, img))
    print("[INFO] Validation split created.")

create_validation_split(train_dir, val_dir)

# === Data transforms ===
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],  # ResNet expects ImageNet normalization
                         [0.229, 0.224, 0.225])
])

# === Dataset and DataLoader ===
train_dataset = datasets.ImageFolder(train_dir, transform=transform)
val_dataset = datasets.ImageFolder(val_dir, transform=transform)
test_dataset = datasets.ImageFolder(test_dir, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

class_names = train_dataset.classes
num_classes = len(class_names)

# === Load Pretrained ResNet18 ===
model = models.resnet18(pretrained=True)

# Freeze early layers
for param in model.parameters():
    param.requires_grad = False

# Replace final layer
model.fc = nn.Sequential(
    nn.Linear(model.fc.in_features, 256),
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(256, num_classes)
)

model = model.to(device)

# === Loss and Optimizer ===
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=0.0005)

# === Training ===
num_epochs = 30
for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)

    # === Validation ===
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    acc = 100 * correct / total
    print(f"Epoch [{epoch+1}/{num_epochs}] - Loss: {avg_loss:.4f}, Val Accuracy: {acc:.2f}%")

# === Testing & Report ===
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for images, labels in test_loader:
        images = images.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())

print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=class_names))

cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
```
